#### Requests & Curl

```python
import requests

headers = {
    'sec-ch-ua': '^\\^Not.A/Brand^\\^;v=^\\^8^\\^, ^\\^Chromium^\\^;v=^\\^114^\\^, ^\\^Google',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
    'Referer': 'https://www.ebooks.com/en-in/subjects/computers/',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
    'Content-Type': 'application/json',
}

params = (
    ('CountryCode', 'IN'),
    ('subjectId', '13'),
)

response = requests.get('https://www.ebooks.com/api/search/subject/', headers=headers, params=params)
```

#### Put everything together

```python
title = []
author = []
publisher = []
publication_year = []
price = []


#title

for result in results_json:
    title.append(result['title'])
```

```python
title
```

```python
# author

for result in results_json:
    author.append(result['authors'][0]['name'])
    
    # publisher
    
    publisher.append(result['publisher'])
    
    # publication year
    
    publication_year.append(result['publication_year'])
    
    # price
    
    price.append(result['price'])
```

```python
publisher
```

```python
publication_year
```

```python
price
```

#### Pandas dataframe

```python
books_df = pd.DataFrame({'Title':title, 'Author':author, 'Publisher':publisher, 
                         'Publication_year':publication_year, 'Price':price})
```

```python
books_df.to_csv('books_info')
```
